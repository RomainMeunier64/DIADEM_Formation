```python
import pandas as pd
import numpy as np

# Affichage amélioré
pd.set_option('display.max_columns', 20)
pd.set_option('display.float_format', '{:.2f}'.format)

print("✅ Pandas prêt :", pd.__version__)
```

## 1. Création d'un DataFrame

Un **DataFrame** est un tableau 2D avec des lignes et des colonnes nommées — comme un tableur Excel, mais en Python.

```python
# Création depuis un dictionnaire
np.random.seed(42)

data = {
    'nom'          : ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Hugo'],
    'age'          : [25, 30, 35, 28, 22, 45, 31, np.nan],
    'ville'        : ['Paris', 'Lyon', 'Paris', 'Bordeaux', 'Lyon', 'Paris', 'Nantes', 'Lyon'],
    'salaire'      : [35000, 42000, 55000, 38000, 29000, 68000, 41000, 37000],
    'anciennete'   : [2, 5, 10, 3, 1, 20, 6, 4],
    'departement'  : ['IT', 'Marketing', 'IT', 'RH', 'Marketing', 'IT', 'RH', None],
    'score_perf'   : [88, 72, 95, 80, 65, 91, 78, 83]
}

df = pd.DataFrame(data)
print(f"DataFrame créé : {df.shape[0]} lignes × {df.shape[1]} colonnes")
df
```

## 2. Exploration des données

```python
# Vue d'ensemble rapide
print("=== FORME ===")
print(f"  {df.shape[0]} lignes, {df.shape[1]} colonnes\n")

print("=== TYPES DES COLONNES ===")
print(df.dtypes)

print("\n=== VALEURS MANQUANTES ===")
missing = df.isnull().sum()
print(missing[missing > 0])
```

```python
# Statistiques descriptives
df.describe()
```

```python
# Premières et dernières lignes
print("=== 3 premières lignes ===")
display(df.head(3))

print("=== 3 dernières lignes ===")
display(df.tail(3))
```

## 3. Sélection et filtrage

```python
# Sélectionner une colonne
print("Salaires :")
print(df['salaire'].values)
```

```python
# Sélectionner plusieurs colonnes
df[['nom', 'ville', 'salaire']]
```

```python
# Filtrer les lignes — employés du département IT
it_employees = df[df['departement'] == 'IT']
print(f"Employés IT : {len(it_employees)}")
it_employees
```

```python
# Filtres combinés — salaire > 40000 ET ville = Paris
filtre = (df['salaire'] > 40000) & (df['ville'] == 'Paris')
df[filtre][['nom', 'ville', 'salaire', 'departement']]
```

```python
# Filtrer avec isin() — villes Lyon ou Nantes
df[df['ville'].isin(['Lyon', 'Nantes'])][['nom', 'ville', 'salaire']]
```

## 4. Nettoyage des données

```python
df_clean = df.copy()  # On travaille sur une copie

# Remplir les valeurs manquantes
df_clean['age'].fillna(df_clean['age'].median(), inplace=True)
df_clean['departement'].fillna('Non défini', inplace=True)

print("Valeurs manquantes après nettoyage :")
print(df_clean.isnull().sum())
```

```python
# Renommer des colonnes
df_clean = df_clean.rename(columns={
    'nom'         : 'Nom',
    'age'         : 'Age',
    'score_perf'  : 'Score'
})

df_clean.head(3)
```

## 5. Transformation et nouvelles colonnes

```python
df2 = df_clean.copy()

# Nouvelle colonne calculée
df2['salaire_mensuel'] = (df2['salaire'] / 12).round(0)

# Catégorisation
df2['niveau_salaire'] = pd.cut(
    df2['salaire'],
    bins=[0, 35000, 50000, 100000],
    labels=['Junior', 'Confirmé', 'Senior']
)

df2[['Nom', 'salaire', 'salaire_mensuel', 'niveau_salaire']]
```

## 6. Agrégations avec groupby()

```python
# Salaire moyen par ville
print("=== Salaire moyen par ville ===")
print(df.groupby('ville')['salaire'].mean().sort_values(ascending=False).round(0))
```

```python
# Agrégation multiple
stats_dept = df.groupby('departement').agg(
    nb_employes   = ('nom', 'count'),
    salaire_moyen = ('salaire', 'mean'),
    salaire_max   = ('salaire', 'max'),
    score_moyen   = ('score_perf', 'mean')
).round(0)

print("=== Statistiques par département ===")
stats_dept
```

## 7. Chargement depuis un fichier CSV (exemple réel)

```python
# Chargement d'un dataset réel depuis seaborn
import seaborn as sns

titanic = sns.load_dataset('titanic')
print(f"Dataset Titanic : {titanic.shape}")
print(f"Colonnes : {list(titanic.columns)}")
titanic.head()
```

```python
# Taux de survie par classe et sexe
titanic.groupby(['pclass', 'sex'])['survived'].mean().round(2).unstack()
```
